# GROUP BY 
-----

```python
# groupby modern and elegant

(emp.groupby(['department_id'])
    .agg(
        Total_salary = ('salary', 'sum'),
        Avg_salary = ('salary', 'mean')
    ).
    sort_values('Total_salary', ascending = False)
)
```

```python
# Avg salary per department
# Highest salary per department
# Median salary per department 

(emp.groupby(['department_id']).
    agg(
        avg_salary = ('salary', 'mean'),
        highest_salary = ('salary', 'max'),
        median_salary = ('salary', 'median')
    )
)
```

```python
# Number of employees per city
(emp.groupby(['city']).
    agg(
        total_employees = ('employee_id', 'count')
    ).
    sort_values('total_employees', ascending = False).
    head(5)
    )
```

```python
# Number of hires per year (extract year from hire_date)
(emp.groupby(['hire_year']).
    agg(
        Total_hired_employees = ('hire_date', 'count')
        ).
    sort_values('Total_hired_employees', ascending = True)
    )
```

# Joins / Merge 

----

```python
# applying merge method to df emp
# merging with dept using 'department_id' 
# using left merge 
# here emp is the leading table all rows from it would be included while only matching rows from dept would be included

(emp.merge(dept, on='department_id', how='left')
     .merge(bonus, on = 'employee_id', how = 'left')
     .head(5)
    )
```

```python
emp[::].sample()
```

```python
dept[::].sample()
```

```python
bonus[::].sample()
```

```python
# Employee + department name
(emp.merge(dept, on = 'department_id', how = 'inner'))[['employee_id', 'Full Name', 'department_id','department_name','budget']]
```

```python
# Employee + bonus (left join  shows NULL/NaN for no bonus)
emp.merge(bonus, on = 'employee_id', how = 'left')[['employee_id','Full Name', 'bonus_amount']]
```

```python
# employee who got bonus
emp.merge(bonus, on = 'employee_id', how = 'right')[['employee_id','Full Name', 'bonus_amount']]
```

```python
# employee who are recoreded in employee table and go bonus
emp.merge(bonus, on = 'employee_id', how = 'inner')[['employee_id','Full Name', 'bonus_amount']]
```

```python
# All departments and their employees (right join on departments)
(emp.merge(dept, on = "department_id", how = 'right').groupby('department_name')
    .agg(
    total_employees = ('employee_id', 'count')
        )
)
```

```python
(emp.merge(dept, on = 'department_id', how = 'outer').
    merge(bonus, on = 'employee_id', how = 'outer')
)[['employee_id', 'Full Name', 'salary', 'performance_score', 'department_name', 'budget', 'bonus_amount']]
```
